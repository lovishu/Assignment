ANS3:In Python, objects are classified as either mutable or immutable based on whether their state (or content) can be changed after they are created. This distinction impacts how objects behave when they are modified.

Mutable Objects:
Definition: Mutable objects are those whose values or state can be changed after creation.
Examples: Lists, dictionaries, sets, and byte arrays.
In this example, the list my_list was changed after creation by modifying the first element.

Immutable Objects:
Definition: Immutable objects are those whose values or state cannot be changed once they are created.
Examples: Integers, floats, strings, tuples, frozensets.

In the case of the list (mutable), modifying one reference affects the other. In the case of the tuple (immutable), modification results in a new object without affecting the original.

Q6: How do conditional statements work in Python? Illustrate with examples?

ANS6:Conditional statements in Python are used to execute code blocks based on whether a given condition is true or false. These conditions typically involve comparison or logical expressions. Python supports several types of conditional statements: if, else, and elif (short for "else if").
1: The if statement executes a block of code if a condition is true.
2: The else statement is used to specify a block of code that will run if the if condition is false.
3: The elif (short for "else if") allows you to check multiple conditions. It is used after an if statement and before an else statement. If the initial if condition is false, Python checks the elif conditions in sequence. If none of the if or elif conditions are true, the else block is executed.

Q7:Describe the different types of loops in Python and their use cases with examples.


ANS7:In Python, loops are used to execute a block of code repeatedly until a certain condition is met. Python supports two types of loops: for loop and while loop. Each loop has different use cases, and they can be used with control statements like break, continue, and pass to modify the loop's behavior.

1:The for loop in Python is used to iterate over a sequence (such as a list, tuple, string, or range). It is used when the number of iterations is known or when you need to iterate over each element in a sequence.
Uses Cases:
Iterating over a collection of items (like lists, tuples, or dictionaries).
Repeating a block of code a specific number of times.
Performing actions on every element of a sequence.

2:The while loop executes a block of code as long as a specified condition is true. It is useful when the number of iterations is not known beforehand, and the loop depends on a condition being met.

Uses Cases:
1:When the number of iterations is not fixed and depends on dynamic conditions.
2:Reading input or processing tasks until a specific event occurs.
3:Continuously polling for events or conditions.

3. Control Statements in Loops
a. break Statement
The break statement is used to terminate the loop prematurely when a certain condition is met. It is used to exit the loop even if the iteration is not complete.
b. continue Statement
The continue statement skips the current iteration and proceeds to the next one. It is used to skip certain values or conditions in the loop without terminating the entire loop.
c. pass Statement
The pass statement does nothing and is used as a placeholder in loops or conditionals where you may have future code but want to avoid syntax errors.
