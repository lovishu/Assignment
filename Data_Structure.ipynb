Q1: Discuss string slicing and provide examples?
ANS1: String slicing in Python refers to extracting a specific portion of a string using a range of indices. It allows you to obtain substrings by specifying a start and end position in the format:

start: The index where slicing begins (inclusive).
end: The index where slicing stops (exclusive).
step: (optional) The step size or increment. If omitted, the default is 1.

Key Points:

Omitting the start defaults to the beginning of the string.
Omitting the end defaults to the end of the string.
Negative indices allow slicing from the end.
The step can be used to control the increment (positive for forward, negative for reverse).

Q4: How to Compare and contrast tuples and lists in the Python?


Ans4: In Python, both tuples and lists are used to store collections of items, but they have key differences in terms of mutability, syntax, and performance. Here's a detailed comparison:

1. Mutability
Lists: Lists are mutable, meaning you can modify them after creation. You can add, remove, or update elements in a list.
Tuples: Tuples are immutable, meaning once a tuple is created, you cannot modify, add, or remove elements.

2: Syntax
Lists: Lists are defined using square brackets [].
Tuples: Tuples are defined using parentheses (). A tuple with a single element requires a comma to differentiate it from a regular expression inside parentheses.

3: Performance
Lists: Lists are slower compared to tuples in certain operations because they are mutable. The flexibility of allowing element modifications and dynamic resizing adds overhead.
Tuples: Tuples are generally faster than lists because they are immutable. Python can optimize tuples better since their size and content don’t change.

4. Use Cases
Lists: Use lists when you need a collection of items that may need to be modified, sorted, or resized during the program's lifecycle.
Tuples: Use tuples when you want a collection of items that should remain constant throughout the program, like fixed data or configuration settings.


Q5: Describe the key features on sets and provide examples of their use in python?

ANS5: Key Features of Sets in Python
1: Unordered Collection:
Sets are unordered, meaning the elements do not have a defined order. This also means that items cannot be accessed by an index.

2: Unique Elements:
A set cannot have duplicate elements. If duplicates are added, they are automatically removed.

3:Mutable:
Sets are mutable, meaning that elements can be added or removed after the set is created, but the elements themselves must be immutable (e.g., strings, numbers, tuples).

4:No Indexing or Slicing:
Unlike lists or tuples, sets do not support indexing or slicing.

5 Set Operations:
Sets support standard operations like union, intersection, difference, and symmetric difference. These operations are useful for comparing data sets.

Q7: Describe how to add, modify and delete items on a dictonary with examples?

1. Adding Items to a Dictionary
To add a new key-value pair to a dictionary, simply assign a value to a new key. If the key already exists, its value will be updated.

my_dict = {"name": "Alice", "age": 25}

2. Modifying Items in a Dictionary
You can modify the value associated with a particular key by reassigning it. If the key exists, it will update the value; if it doesn’t exist, a new key-value pair is added.


3. Deleting Items from a Dictionary
You can delete items from a dictionary using several methods:

A. Using del:
Deletes the key-value pair from the dictionary.
Raises a KeyError if the key is not found.

B. Using pop():
Removes the key and returns the corresponding value.

Raises a KeyError if the key is not found, but you can provide a default return value to avoid this.

C. Using popitem():
Removes and returns the last key-value pair added to the dictionary.
Raises a KeyError if the dictionary is empty.

D. Using clear():
Removes all key-value pairs, leaving an empty dictionary.




Q8: Discusses the importance of dictionary keys being immutable and provide exmaples?

ANS8: In Python, dictionary keys must be immutable because dictionaries use a technique called hashing to store and retrieve values efficiently. The key's immutability ensures that once it's used in a dictionary, its hash value won't change, maintaining the dictionary's integrity.

Why must dictionary keys be immutable?
1: Hashing mechanism: Python dictionaries rely on a hash table internally. Each key is passed through a hash function to generate a hash value, which is used to place the key-value pair in the dictionary. For this to work consistently, the key’s hash value must remain constant. If keys could be mutable, modifying a key after it’s been hashed would disrupt the lookup process, as the new hash wouldn’t correspond to the original one.

2: Data integrity: If dictionary keys were mutable, changing a key’s value would make it difficult to ensure that the dictionary functions correctly, leading to unpredictable behavior and possible data loss. The dictionary could lose track of the entry because the key is no longer found at its original hashed position.

Examples of Immutable and Mutable Types

Immutable types (allowed as dictionary keys):

Strings: "name", "ID", etc.
Numbers: 1, 3.14, etc.
Tuples: (1, 2, 3), ("x", "y") (as long as all elements in the tuple are also immutable)

Mutable types (not allowed as dictionary keys):

Lists: [1, 2, 3], ["a", "b", "c"]
Sets: {1, 2, 3}
Dictionaries: {1: "a", 2: "b"}

Valid dictionary with immutable keys:

my_dict = {
               "name": "Alice",
               101: "Student ID",     
               (1, 2, 3): "Tuple as key"
}
