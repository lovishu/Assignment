Ans6.Handling Categorical Variables in Machine Learning
Categorical variables need to be converted into numerical form before being used in most machine learning algorithms.

âœ… Common Techniques to Handle Categorical Variables:
1. Label Encoding
Converts categories into integer labels.

Red â†’ 0, Green â†’ 1, Blue â†’ 2
