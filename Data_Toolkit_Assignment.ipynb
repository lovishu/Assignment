Q3  5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a

Ans3. import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

# Generate synthetic data with correlation
size = 100
x1 = np.random.normal(0, 1, size)
x2 = 0.8 * x1 + np.random.normal(0, 0.2, size)  # correlated with x1
x3 = np.random.normal(0, 1, size)               # not correlated
x4 = x1 - x2 + np.random.normal(0, 0.1, size)   # mix of x1 and x2

# Create DataFrame
df = pd.DataFrame({
    'Feature1': x1,
    'Feature2': x2,
    'Feature3': x3,
    'Feature4': x4
})

# Compute correlation matrix
corr_matrix = df.corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Correlation Matrix Heatmap')
plt.show()

Q1. 1. Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a three dimensional space.

Ans1.

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate synthetic 3D data
np.random.seed(0)
x = np.random.rand(100) * 10
y = np.random.rand(100) * 10
z = np.random.rand(100) * 10

# Create a 3D scatter plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the points
scatter = ax.scatter(x, y, z, c=z, cmap='viridis', marker='o')

# Add labels and title
ax.set_title('3D Scatter Plot of Data Points')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

# Show color bar and plot
fig.colorbar(scatter, ax=ax, label='Z value')
plt.show()

Bokeh Assignment


 Q1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

 Ans1.

from bokeh.plotting import figure, show, output_file
import numpy as np

# Generate data
x = np.linspace(0, 10, 500)
y = np.sin(x)

# Output to an HTML file
output_file("sine_wave.html")

# Create a Bokeh figure
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=700, height=400)

# Add a line renderer
p.line(x, y, legend_label="sin(x)", line_width=2, line_color="navy")

# Show the plot
show(p)
