Q6. Explain the concept of generators in Python and how they are defined.

Ans6. A generator is a special type of iterator that yields values one at a time, on demand, using the yield keyword instead of return. They are memory-efficient and ideal for working with large datasets or infinite sequences.

def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1

Q1.Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list

    # Use a list comprehension to filter even numbers and sum them
    return sum(x for x in numbers if x % 2 == 0)

Q4.  Write a Python function that checks if a given number is prime or not from 1 to 200.

Ans4. def is_prime(number):
    if number <= 1:
        return False
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            return False
    return True

# Example usage:
for num in range(1, 201):
    if is_prime(num):
        print(num, "is prime")

Q5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

Ans5. class FibonacciIterator:
    def __init__(self, limit):
        self.limit = limit  # The upper limit for the Fibonacci sequence
        self.a, self.b = 0, 1  # Starting values of the Fibonacci sequence

    def __iter__(self):
        return self

    def __next__(self):
        if self.a > self.limit:
            raise StopIteration  # Stop when the value exceeds the specified limit
        else:
            current = self.a
            self.a, self.b = self.b, self.a + self.b  # Update the Fibonacci sequence
            return current

# Example usage:
fibonacci_sequence = FibonacciIterator(100)  # Generate Fibonacci sequence up to 100
for num in fibonacci_sequence:
    print(num)

Q6.  6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

Ans6. def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

# Example usage:
for power in powers_of_two(5):
    print(power)


Q7.Implement a generator function that reads a file line by line and yields each line as a string.

Ans7.def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()  # Strip to remove any leading/trailing whitespace or newline characters

# Example usage:
file_path = 'example.txt'  # Replace with the path to your file
for line in read_file_line_by_line(file_path):
    print(line)

Q9.Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

Ans9. def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# List of temperatures in Celsius

# Using map() to convert the list of Celsius temperatures to Fahrenheit
