Ans3.def is_palindrome(s):
    # Remove spaces and convert to lowercase for uniformity
    cleaned = ''.join(char.lower() for char in s if char.isalnum())
    return cleaned == cleaned[::-1]

    positions = []
    start = 0
    while True:
        index = text.find(substring, start)
        if index == -1:

        positions.append(index)

    return positions

Ans6. def compress_string(s):
    if not s:
        return ""

    compressed = []
    count = 1

    for i in range(1, len(s)):
        if s[i] == s[i - 1]:
            count += 1
        else:
            compressed.append(s[i - 1] + str(count))
            count = 1

    # Append the last character group
    compressed.append(s[-1] + str(count))

    compressed_str = ''.join(compressed)

    # Return the compressed string only if it's shorter
    return compressed_str if len(compressed_str) < len(s) else s

# Example usage
input_str = "aabcccccaaa"
output = compress_string(input_str)
print("Compressed string:", output)

Q14.  Write a code to reverse a list in-place without using any built-in reverse functions

Ans14. def reverse_list(lst):
    left, right = 0, len(lst) - 1
    while left < right:
        # Swap the elements at left and right positions
        lst[left], lst[right] = lst[right], lst[left]
        left += 1
        right -= 1

# Example usage
input_list = [1, 2, 3, 4, 5]
reverse_list(input_list)
print("Reversed list:", input_list)

Q17.  Write a code to merge two sorted lists into a single sorted list

Ans17. def merge_sorted_lists(list1, list2):
    # Initialize pointers for both lists
    i, j = 0, 0
    merged_list = []

    # Merge the two lists
    while i < len(list1) and j < len(list2):
        if list1[i] < list2[j]:
            merged_list.append(list1[i])
            i += 1
        else:
            merged_list.append(list2[j])
            j += 1

    # Append any remaining elements from either list
    merged_list.extend(list1[i:])
    merged_list.extend(list2[j:])

    return merged_list

# Example usage
list1 = [1, 3, 5, 7]
list2 = [2, 4, 6, 8]
result = merge_sorted_lists(list1, list2)
print("Merged sorted list:", result)

Q20. Write a code to shuffle a given list randomly without using any built-in shuffle functions

Ans20. import random

def shuffle_list(lst):
    # Get the length of the list
    n = len(lst)
    
    # Shuffle the list by swapping elements randomly
    for i in range(n):
        # Randomly select an index from the remaining elements
        j = random.randint(i, n-1)
        
        # Swap the elements at indices i and j
        lst[i], lst[j] = lst[j], lst[i]

# Example usage
input_list = [1, 2, 3, 4, 5]
shuffle_list(input_list)
print("Shuffled list:", input_list)

def merge_and_add_dicts(dict1, dict2):
    # Create a new dictionary to store the result
    merged_dict = dict1.copy()  # Copy dict1 to preserve it
    
    # Iterate over dict2 and add values to merged_dict
    for key, value in dict2.items():
        if key in merged_dict:
            # If the key is already in dict1, add the value from dict2 to dict1's value
            merged_dict[key] += value
        else:
            # If the key is not in dict1, simply add it to merged_dict

def get_nested_value(dictionary, keys):
    # Loop through the keys and access the corresponding value in the dictionary
    for key in keys:
        # If the key doesn't exist, return None
        if key not in dictionary:
            return None
        dictionary = dictionary[key]  # Update dictionary to access nested level
    return dictionary

 Write a code that inverts a dictionary, swapping keys and values. Ensure that the inverted dictionary correctly handles cases where multiple keys have the same value by storing the keys as a list in the inverted dic

Ans35. def invert_dict(dictionary):
    inverted_dict = {}
    
    for key, value in dictionary.items():
        # If the value is already a key in the inverted dictionary, append the current key to the list
        if value in inverted_dict:
            inverted_dict[value].append(key)
        else:
            # If the value is not a key, create a new list with the current key
            inverted_dict[value] = [key]
    
    return inverted_dict

# Example usage
input_dict = eval(input("Enter the dictionary (e.g., {'a': 1, 'b': 2, 'c': 1}): "))

# Invert the dictionary
inverted_result = invert_dict(input_dict)

# Print the inverted dictionary
print("Inverted dictionary:", inverted_result)
