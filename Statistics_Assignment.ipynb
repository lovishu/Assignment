Q3. 3. Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?


Ans3.3. Concept of Dispersion
Dispersion refers to the extent to which data values vary or spread out around a central value (like the mean). It helps in understanding the distribution and consistency of the data.

A dataset with high dispersion has values spread widely, while a dataset with low dispersion has values close to the mean.

Key Measures of Dispersion
Range: Difference between the maximum and minimum values.

Variance: Average of the squared deviations from the mean.

Standard Deviation: Square root of variance; shows average distance from the mean in original units.

Q6. Explain the concept of skewness and its types. How does skewness affect the interpretation of data?

Ans6.  Skewness refers to the asymmetry in the distribution of data. It shows whether the data is symmetrically distributed or leans more to one side.

If a distribution is symmetric, the mean = median = mode.
If not symmetric, the distribution is skewed, affecting the interpretation of central values.


Types of Skewness

Shape: Bell curve (e.g., normal distribution)

2. Positive Skew (Right-Skewed)
Tail extends to the right (towards higher values)

3. Negative Skew (Left-Skewed)
Tail extends to the left (towards lower values)




Q7.What is the interquartile range (IQR), and how is it used to detect outliers?

Ans7. The Interquartile Range (IQR) measures the spread of the middle 50% of a dataset. It is the difference between the third quartile (Q3) and the first quartile (Q1):

Q1 (25th percentile): The value below which 25% of the data fall

Q3 (75th percentile): The value below which 75% of the data fall



How IQR is Used to Detect Outliers
Outliers are data points that fall significantly outside the range of most of the data. They can be detected using the 1.5 √ó IQR Rule:

Outlier Detection Rule:
Lower bound = Q1 ‚àí 1.5 √ó IQR

Upper bound = Q3 + 1.5 √ó IQR

Any data point outside these bounds is considered an outlier.
