Q2: Describe the role of predefined keywords in Python and provide examples of how they are used in a  program?

ANS2: In Python, predefined keywords (also known as reserved words) are part of the language's syntax and have specific meanings. These keywords cannot be used as identifiers (like variable names, function names, or any other names) because they are reserved for defining the structure and control flow of the program.

Role of Predefined Keywords:
1:Control the Flow of a Program: Keywords such as if, else, while, and for are used to create conditional statements and loops.
2:Define Functions, Classes, and Variables: Keywords like def, class, and global help in defining functions, classes, and controlling the scope of variables.
3:Handle Exceptions: Keywords like try, except, and finally are used for exception handling.
4:Work with Data Structures: Keywords like True, False, None represent Boolean values and null values, respectively.
5:Memory Management: Keywords such as del are used to free up memory by removing references to objects.
